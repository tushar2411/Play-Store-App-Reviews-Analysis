#*In Rating column we have some null values. To remove these values we'll calculate median of this column by creating function*

```python
def cal_median(counts):
  return counts.fillna(counts.median)
```

```python
playdataF.Rating = playdataF['Rating'].transform(cal_median)
```

#*Now we will apply this function for same the 'Translated_Review,Sentiment,Sentiment_Polarity,Sentiment_Subjectivity'*

```python
playdataF.Translated_Review = playdataF['Translated_Review'].transform(cal_median)
```

```python
playdataF.Sentiment = playdataF['Sentiment'].transform(cal_median)
```

```python
playdataF.Sentiment_Polarity = playdataF['Sentiment_Polarity'].transform(cal_median)
```

```python
playdataF.Sentiment_Subjectivity = playdataF['Sentiment_Subjectivity'].transform(cal_median)
```

```python
playdataF.isnull().sum()
```

#*We can also see that above information about data there is a Price column which has all numeric float values plus there is $ sign we will remove it and convert into float values and also*

```python
playdataF['Price'].unique()
```

```python
playdataF['Price'] = playdataF['Price'].astype('float')
```

```python
for i in range(0,len(playdataF['Price'])):
  if '$' in playdataF.loc[i,'Price']:
    playdataF.loc[i,'Price'] = playdataF.loc[i,'Price'][1:]
  playdataF.loc[i,'Price'] =  int(playdataF.loc[i,'Price'])
```
